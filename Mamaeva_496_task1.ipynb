$$l_t = \alpha (y_t - s_{t-p}) + (1-\alpha) (l_{t-1} + b_{t-1} ) \color{red}{=} l_{t-1} + b_{t-1} + \alpha e_t;$$
    $$ b_t = \beta (l_{t} - l_{t-1} ) + (1-\beta) b_{t-1} \color{red}{=} b_{t-1} + \alpha\beta e_t;$$
    $$ s_t = \gamma (y_t-l_t) + (1-\gamma) s_{t-p} \color{red}{=} s_{t-p} + \gamma(1-\alpha)e_t.$$

*Решение:*


$\hat{y}_t = (l_{t - 1} + b_{t - 1}) + s_{t - p}, e_t := y_t - \hat{y}_t = \underline{y_t - l_{t - 1} - b_{t - 1} - s_{t - p}}$

$l_t = \alpha (y_t - s_{t-p}) + (1 - \alpha) (l_{t - 1} + b_{t - 1}) = \alpha ({y_t - s_{t - p} - l_{t - 1} - b_{t - 1}}) + l_{t - 1} + b_{t - 1} = \alpha e_t + l_{t - 1} + b_{t - 1}$

$b_t = \beta(l_t - l_{t - 1}) + (1 - \beta) b_{t - 1} = \beta (l_t - l_{t - 1} - b_{t - 1}) + b_{t - 1} = \beta(l_{t - 1} + b_{t - 1} + \alpha e_t - l_{t - 1} - b_{t - 1}) + b_{t - 1} = \alpha\beta e_t + b_{t - 1}$

$s_t = \gamma (y_t - l_t) + (1 - \gamma) s_{t - p} = \gamma (y_t - l_t - s_{t - p}) + s_{t - p} = \gamma ({y_t - l_{t - 1} - b_{t - 1} - s_{t - p} - \alpha e_t}) + s_{t - p} = \gamma(e_t - \alpha e_t) + s_{t - p} = \gamma (1 - \alpha) e_t + s_{t - p}$


**Вопрос 4**: Выведите формулу расчёта скрытых компонент в форме корректировки прогноза (correction error form) временного ряда в модели ЭС с линейным затухающим трендом (WS with additive damped trend):

$$\hat{y}_{t+d|t} = l_t + \left(\phi + \phi^2 + \dots + \phi^{d}\right) b_t.$$
   
Скрытые компоненты ВР согласно модели в рекуррентной форме записываются:
    $$ l_{t}       = \alpha y_t + \left(1-\alpha\right) \left(l_{t-1} +\phi b_{t-1}\right),$$
    $$ b_t         = \beta \left(l_t - l_{t-1}\right) + \left(1-\beta\right)\phi b_{t-1}.$$

*Решение:*

$e_t = y_t - \hat y_{t ~|~ t - 1} = y_t - l_{t - 1} - \phi b_{t - 1}$

$b_t = \beta (l_t - l_{t - 1}) + (1 - \beta) \phi b_{t - 1} = \beta (l_t - \phi b_{t - 1} - l_{t - 1}) + \phi b_{t - 1} = \beta \alpha e_t + \phi b_{t - 1}$
$l_t = \alpha y_t + (1 - \alpha) (l_{t - 1} + \phi b_{t - 1}) = \alpha ({y_t - l_{t - 1} - \phi b_{t - 1}}) + l_{t - 1} + \phi b_{t - 1} = \alpha e_t + l_{t - 1} + \phi b_{t - 1}$

```python
# Start with this code
###################### Winters Exponential Smoothing #########################
# x <array Tx1>- time series, 
# h <scalar> - forecasting delay
# Params <dict> - dictionary with 
#    alpha <scalar in [0,1]> - smoothing parameter
#    delts <scalar in [0,1]> - seasonality smoothing parameter

def WintersExponentialSmoothing(x, h, Params):
    T = len(x)
    alpha = Params['alpha']
    delta = Params['delta']
    p = Params['seasonality_period']
    
    FORECAST = [np.NaN]*(T + h)
    
    l = np.NaN  
    s = [np.NaN] * p  
    
    for cntr in range(T):
        if not math.isnan(x[cntr]):
            if math.isnan(l):
                l = x[cntr]  
 
            if math.isnan(s[cntr % p]):
                s[cntr % p] = 1  
            
            lp = l
            l =  alpha * (x[cntr] - s[(cntr - p) % p]) + (1 - alpha) * l  
            s[cntr % p] = delta * (x[cntr] - lp) + (1 - delta) * s[(cntr - p) % p] 
            
        FORECAST[cntr+h] = l + s[(cntr - p + h)  % p]
    return FORECAST
```

```python
def BuildForecast(h, ts, AlgName, AlgTitle, ParamsArray, step='D'):
    FRC_TS = dict()
    for p in ParamsArray:
        frc_horizon = pd.date_range(ts.index[-1], periods=h+1, freq=step)[1:]
        frc_ts = pd.DataFrame(index = ts.index.append(frc_horizon), columns = ts.columns)
        
        for cntr in ts.columns:
            frc_ts[cntr] = eval(AlgName)(ts[cntr], h, p)
        
        FRC_TS['%s %s' % (AlgTitle, p)] = frc_ts
    return FRC_TS
```

```python
ALPHA = np.arange(0, 1, 0.01)
DELTA = np.arange(0, 1, 0.01)
ESParamsArray = [{'alpha':alpha, 'delta':delta, 'seasonality_period':12} for alpha in ALPHA for delta in DELTA]
FRC_WAGE = BuildForecast(h=1, ts=wage, AlgName = 'WintersExponentialSmoothing', AlgTitle='WES',ParamsArray = ESParamsArray)

# compare ES parameters
QualityStr = pd.DataFrame(index = wage.columns, columns = FRC_WAGE.keys())

ix = wage.loc['1998-09-01':'2015-01-01'].index
for param_cntr in sorted(QualityStr.columns):
    frc_wage = FRC_WAGE[param_cntr]
    QualityStr[param_cntr],_ = qualityMACAPE(wage.loc[ix], frc_wage.loc[ix])

QualityStr[QualityStr.columns].mean().sort_values()[:10]
```

```python
###################### Theil-Wage model #########################
# x <array Tx1>- time series, 
# h <scalar> - forecasting delay
# Params <dict> - dictionary with 
#    alpha, beta, gamma <scalar in [0,1]> - parameters
#    seasonality_period


def TheilWage(x, h, Params):
    T = len(x)
    alpha = Params['alpha']
    beta = Params['beta']
    gamma = Params['gamma']
    p = Params['seasonality_period']
    
    FORECAST = [np.NaN]*(T+h)
    
    
    l = np.NaN  
    b = np.NaN 
    s = [np.NaN] * p  
    
    for cntr in range(T):
        if not math.isnan(x[cntr]):
            if math.isnan(l):
                l = x[cntr]  
 
            if math.isnan(b):
                b = 1 
 
            if math.isnan(s[cntr % p]):
                s[cntr % p] = 1 
            
            lp = l
            l =  alpha * (x[cntr] - s[(cntr - p) % p]) + (1 - alpha) * (l + b)  
            b = beta * (l - lp) + (1 - beta) * b 
            s[cntr % p] = gamma * (x[cntr] - l) + (1 - gamma) * s[(cntr - p) % p]  
            
        FORECAST[cntr + h] = abs(l + b * h * s[(cntr - p + h) % p])
    return FORECAST
```
